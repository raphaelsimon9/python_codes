```python
df['State'].dropna(inplace= True)
```

```python
df['Company'].value_counts()
```

```python
df['Company'] = df['Company'].str.lower().str.title()
```

```python
df['Company public response'].value_counts()
```

```python
df['Company public response'].mode()
```

```python
df['Company public response'] = df['Company public response'].fillna(df['Company public response'].mode()[0]) # mode()[0] ensures that the first mode result is returned and used to fill the nulls
```

```python
df['Consumer complaint narrative'].value_counts()
```

```python
df['Consumer complaint narrative'] = df['Consumer complaint narrative'].fillna('No Complaint Given')
```

```python
df.dropna(subset=['State'], inplace= True)
```

```python
df.isnull().sum()
```

```python
df['Sub-issue'] = df['Sub-issue'].fillna('No Sub-Issue Given')
```

```python
df.columns
```

```python
df['Product']= df['Product'].str.title()
df['Product'].value_counts()
```

```python
df["Product"] = (
    df["Product"]
    .astype(str)
    .str.strip()                 # remove extra spaces
    .str.lower()                 # lowercase
    .str.replace(r"\s+", " ", regex=True)   # compress multiple spaces
)
```
